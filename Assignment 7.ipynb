Arithmetic Operators in Pytho


Addition(+)

```python
a = 20
b = 30
result = a + b
print(f"{a} + {b} = {result}")
```

Subtraction (-)

```python
a = 50
b = 30
result= a - b
print(f"{a} - {b} = {result}")
```

Multiplication (-)


```python
a = 30
b = 20
result = a * b
print(f"{a} * {b} = {result}")
```

Division (/)


```python
a = 10
b = 3
result = a / b
print(f"{a} / {b} = {result}")
```

Floor Division (//)


```python
a = 10
b = 3
result = a // b
print(f"{a} // {b} = {result}")
```

Modulus (%)


```python
a = 10
b = 3
result = a % b
print(f"{a} % {b} = {result}")
```

Exponentiation(**)

```python
a = 10
b = 3
result = a ** b
print(f"{a} ** {b} = {result}")
```
